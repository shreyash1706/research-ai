```python
import os
import pandas as pd
from transformers import AutoTokenizer
import matplotlib.pyplot as plt
```

```python
file= "metadata/arxiv_metadata.csv"

df = pd.read_csv(file,on_bad_lines='skip')
df.head()
```

```python
df = df.dropna(axis=0)
```

```python
df
```

```python
df = df[df['summary'].str.count(r'\\') <= 5]

# Optional: verify
print(f"Rows remaining: {len(df)}")
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    "mistralai/Mistral-7B-v0.1",
    use_fast=True
)
```

```python
from tqdm import tqdm
tqdm.pandas()

def count_tokens(text):
    if not isinstance(text, str) or text.strip() == "":
        return 0
    return len(tokenizer(text, add_special_tokens=False)["input_ids"])

df["token_count"] = df["summary"].progress_apply(count_tokens)
```

```python
df["token_count"].describe()
```

```python


plt.figure(figsize=(8, 5))
plt.hist(df["token_count"], bins=100)
plt.xlabel("Token count")
plt.ylabel("Number of abstracts")
plt.title("Abstract Token Length Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["token_count"], bins=100, log=True)
plt.xlabel("Token count")
plt.ylabel("Log frequency")
plt.title("Abstract Token Length Distribution (log scale)")
plt.show()
```

```python
df["published"] = pd.to_datetime(
    df["published"],
    format="mixed",
    errors="coerce",
    utc=True
)
```

```python
df["year"] = df["published"].dt.year
```

```python
CORE_CATS = {
    "cs.AI": 3,
    "cs.LG": 3,
    "stat.ML": 3,
    "cs.CL": 2,
    "cs.CV": 2,
    "cs.NE": 2,
}

DEFAULT_WEIGHT = 1
```

```python
def category_score(cat_str):
    if not isinstance(cat_str, str):
        return DEFAULT_WEIGHT
    
    score = 0
    for cat, w in CORE_CATS.items():
        if cat in cat_str:
            score += w
    
    return score if score > 0 else DEFAULT_WEIGHT


df["cat_score"] = df["categories"].apply(category_score)
```

```python
def length_bucket(n):
    if n < 150:
        return "short"
    elif n <= 300:
        return "medium"
    else:
        return "long"

df["length_bucket"] = df["token_count"].apply(length_bucket)
```

```python
SAMPLE_SIZE = 30000

df_sampled = df.sample(
    n=SAMPLE_SIZE,
    weights=df["cat_score"],
    random_state=42
)
```

```python
print(df_sampled["categories"].str.contains("cs.AI").mean(),
df_sampled["categories"].str.contains("cs.LG").mean(),
df_sampled["categories"].str.contains("stat.ML").mean())
```
